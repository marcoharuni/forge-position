## 3. Correct prefill and decode positions

```python
batch_size = 1
prompt_len = 5
prefill = build_position_ids_for_prefill(batch_size, prompt_len)
decode = build_position_ids_for_decode(batch_size, past_len=prompt_len)
print('prefill:', prefill.tolist())
print('first decode:', decode.tolist())
```

## 4. Reset-position bug

```python
wrong_decode = build_position_ids_for_decode(batch_size, past_len=0)
print('wrong first decode:', wrong_decode.tolist())
print('bug: next token reuses RoPE phase 0 instead of phase 5')
```

## 5. Left-padding bug

```python
wrong_left_pad = torch.tensor([[0, 0, 1, 2, 3]])
correct_left_pad = torch.tensor([[0, 0, 0, 1, 2]])
print('wrong left-pad positions:', wrong_left_pad.tolist())
print('one valid convention:', correct_left_pad.tolist())
```
